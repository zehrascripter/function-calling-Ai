```python
@tool
def info(query: str) -> str:
    """
    Retrieve information about the student information according to their key.
    """
    if query in students:
        return f"The data for {query} is: {students[query]}"
    else:
        return f"No data available for {query}. Please try again."
```

```python
tools = [info]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```

```python
user_input = "who is PIAIC00001 show all details"
responce = agent.invoke(user_input)
print(responce)
```
